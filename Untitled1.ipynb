```python
import yfinance as yf

# Fetch Tesla stock data
tesla = yf.Ticker("TSLA")
tesla_stock_data = tesla.history(period="max")

# Display the first few rows
tesla_stock_data.head()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL for Tesla's quarterly revenue data
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"

# Send HTTP request
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# Initialize an empty DataFrame
tesla_revenue = pd.DataFrame(columns=["Date", "Revenue"])

# Locate the revenue table and extract data
for table in soup.find_all("table"):
    if "Tesla Quarterly Revenue" in str(table):
        rows = table.find_all("tr")
        for row in rows[1:]:
            cols = row.find_all("td")
            if len(cols) == 2:
                date = cols[0].text.strip()
                revenue = cols[1].text.strip().replace("$", "").replace(",", "")
                if revenue != "":
                    tesla_revenue = tesla_revenue.append({"Date": date, "Revenue": float(revenue)}, ignore_index=True)

# Display the first few rows
tesla_revenue.head()
```

```python
# Fetch GameStop stock data
gamestop = yf.Ticker("GME")
gamestop_stock_data = gamestop.history(period="max")

# Display the first few rows
gamestop_stock_data.head()
```

```python
# URL for GameStop's quarterly revenue data
url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"

# Send HTTP request
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# Initialize an empty DataFrame
gamestop_revenue = pd.DataFrame(columns=["Date", "Revenue"])

# Locate the revenue table and extract data
for table in soup.find_all("table"):
    if "GameStop Quarterly Revenue" in str(table):
        rows = table.find_all("tr")
        for row in rows[1:]:
            cols = row.find_all("td")
            if len(cols) == 2:
                date = cols[0].text.strip()
                revenue = cols[1].text.strip().replace("$", "").replace(",", "")
                if revenue != "":
                    gamestop_revenue = gamestop_revenue.append({"Date": date, "Revenue": float(revenue)}, ignore_index=True)

# Display the first few rows
gamestop_revenue.head()
```

```python
import matplotlib.pyplot as plt

def plot_dashboard(stock_data, revenue_data, company_name):
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))

    # Plot stock price
    ax1.plot(stock_data.index, stock_data['Close'], label='Stock Price')
    ax1.set_title(f'{company_name} Stock Price')
    ax1.set_xlabel('Date')
    ax1.set_ylabel('Price (USD)')

    # Plot revenue
    ax2.bar(revenue_data['Date'], revenue_data['Revenue'], label='Revenue', color='orange')
    ax2.set_title(f'{company_name} Quarterly Revenue')
    ax2.set_xlabel('Date')
    ax2.set_ylabel('Revenue (USD)')
    ax2.tick_params(axis='x', rotation=45)

    plt.tight_layout()
    plt.show()

# Plot Tesla dashboard
plot_dashboard(tesla_stock_data, tesla_revenue, "Tesla")
```

```python
# Plot GameStop dashboard
plot_dashboard(gamestop_stock_data, gamestop_revenue, "GameStop")
```
